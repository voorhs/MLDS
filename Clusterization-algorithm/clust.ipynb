```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns
from clust import KMeans
from clust import elbow
```

```python
X, y = make_blobs(n_samples=2000, n_features=2, centers=4, random_state=0)
```

```python
m = KMeans(X, n_clusters=4, n_features=2).fit(heur='sample', n_start=3)

fig, ax = plt.subplots(1, 2)

ax[0].scatter(X[:, 0], X[:, 1], c=m.labels, s=5)
ax[0].scatter(m.centroids[:, 0], m.centroids[:, 1], s=120)

ax[1].scatter(X[:, 0], X[:, 1], c=y, s=5)
ax[1].scatter(m.centroids[:, 0], m.centroids[:, 1], s=120)
```

```python
solve = elbow(X, n_features=2)
```

```python
plt.plot(solve.variances_for_k)
solve.choose_k()
```
