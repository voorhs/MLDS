```python
import numpy as np
from clust import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=2, random_state=0)
```

```python
m = KMeans(2).fit(X, eur='sample')

x, y = np.hsplit(X, 2)
plt.scatter(x, y, c=m.labels)
c1, c2 = np.hsplit(m.centroids, 2)
plt.scatter(c1, c2)
```
