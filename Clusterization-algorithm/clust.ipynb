```python
import numpy as np
from test import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns
```

```python
X, y = make_blobs(n_samples=1000, n_features=2, centers=5, random_state=0)
```

```python
m = KMeans(X, 5).fit(heur='sample')

fig, ax = plt.subplots(1, 2)
ax[0].scatter(X[:, 0], X[:, 1], c=m.labels)
ax[0].scatter(m.centroids[:, 0], m.centroids[:, 1], s=200)
m.centroids

ax[1].scatter(X[:, 0], X[:, 1], c=y)
ax[1].scatter(m.centroids[:, 0], m.centroids[:, 1], s=200)
```
