```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
from scipy.stats import uniform
from scipy.stats import expon
from scipy.stats import geom
```

```python
size = 1000
N1 = norm.rvs(size=size) * 2 + 5
N2 = norm.rvs(size=size) * 0.5 + 5
U = uniform.rvs(size=size) * 8 + 1
E = expon.rvs(scale=1, size=size)
G = geom.rvs(p=0.3, size=size)
```

```python
fig, ax = plt.subplots(1, figsize=(16,7))
ax.set_xticklabels(['$N(5,2^2)$', '$N(5, 0.5^2)$', '$U[1,9]$', '$Exp(1)$', '$Geom(0.3)$'], fontdict={'fontsize': 15})

ax.boxplot([N1, N2, U, E, G])
plt.show()
```
