```python
import pandas as pd
```

```python
df = pd.read_table('L1S5T2')
df.head()
```

```python
arts = df[df['School_Type'] == 'Lib Arts']
arts.head()
```

```python
univ = df[df['School_Type'] == 'Univ']
univ.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20,10))

ax[0, 0].boxplot([univ['SAT'],          arts['SAT']],           labels=['Univ', 'Arts'])
ax[0, 1].boxplot([univ['Acceptance'],   arts['Acceptance']],    labels=['Univ', 'Arts'])
ax[0, 2].boxplot([univ['$/Student'],    arts['$/Student']],     labels=['Univ', 'Arts'])
ax[1, 0].boxplot([univ['Top 10%'],      arts['Top 10%']],       labels=['Univ', 'Arts'])
ax[1, 1].boxplot([univ['%PhD'],         arts['%PhD']],          labels=['Univ', 'Arts'])
ax[1, 2].boxplot([univ['Grad%'],        arts['Grad%']],         labels=['Univ', 'Arts'])

ax[0, 0].set_title('SAT')
ax[0, 1].set_title('Acceptance')
ax[0, 2].set_title('$/Student')
ax[1, 0].set_title('Top 10%')
ax[1, 1].set_title('%PhD')
ax[1, 2].set_title('Grad%')

plt.show()
```

```python
fig, ax = plt.subplots(6, 2, figsize=(8,22))

bins = 12

ax[0, 0].hist(univ['SAT'], bins)
ax[0, 1].hist(arts['SAT'], bins)

ax[1, 0].hist(univ['Acceptance'], bins)
ax[1, 1].hist(arts['Acceptance'], bins)

ax[2, 0].hist(univ['$/Student'], bins)
ax[2, 1].hist(arts['$/Student'], bins)

ax[3, 0].hist(univ['Top 10%'], bins)
ax[3, 1].hist(arts['Top 10%'], bins)

ax[4, 0].hist(univ['%PhD'], bins)
ax[4, 1].hist(arts['%PhD'], bins)

ax[5, 0].hist(univ['Grad%'], bins)
ax[5, 1].hist(arts['Grad%'], bins)

ax[0, 0].set_ylabel('SAT')
ax[0, 0].set_title('Univ')
ax[0, 1].set_title('Arts')

ax[1, 0].set_ylabel('Acceptance')
ax[2, 0].set_ylabel('$/Student')
ax[3, 0].set_ylabel('Top 10%')
ax[4, 0].set_ylabel('%PhD')
ax[5, 0].set_ylabel('Grad%')

plt.show()
```
